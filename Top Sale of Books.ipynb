```python
def showpage(url,page):
    html = requests.get(url, headers = headers).text
    soup = BeautifulSoup(html,"lxml")
    res = soup.find_all("div",{"class":"mod type02_m012 clearfix"})[0]
    items = res.select(".item")
#     item = items[0]        
    for item in items:
        msg = item.select(".msg")[0]
        src = item.select("a img")[0]["src"]
        sd = datetime.strptime('2022/2/1',"%Y/%m/%d")
        ed = datetime.strptime('2022/4/30',"%Y/%m/%d")
        mid_date = msg.find("span").text.split("：")[-1]
        date = datetime.strptime(mid_date,"%Y/%m/%d")
        if sd < date < ed:
            title = msg.select("a")[0].text
            print("書名：",title)
            author = msg.select("a")[1].text
            print("作者：",author)
            publish = msg.select("a")[2].text
            print("出版社：",publish)
            print("出版日期：",date)
            pricebox = item.select(".price .set2")[0]
            onsale = pricebox.select("strong")[1].text
            print("優惠價：",onsale)
            rate = int(twobyte(pricebox.select("strong")[0].text))/100
            price = round(int(onsale)/rate)
            print("定價：",price)
            content = msg.select(".txt_cont")[0].text.replace(" ","").strip()
            print("內容：",content)
            listdata = [title,author,publish,date,price,onsale,content]
            list1.append(listdata)
            print("-"*50)
            print()
```

```python
def showsdata(url,page):
    html = requests.get(url, headers = headers).text
    soup = BeautifulSoup(html,"lxml")
    res = soup.find_all("div",{"class":"mod type02_m012 clearfix"})[0]
    items = res.select(".item")
    c = 0
    pd = {}
    listpd = []
    C = []
    for item in items:
        msg = item.select(".msg")[0]
        sd = datetime.strptime('2022/2/1',"%Y/%m/%d")
        ed = datetime.strptime('2022/4/30',"%Y/%m/%d")
        mid_date = msg.find("span").text.split("：")[-1]
        date = datetime.strptime(mid_date,"%Y/%m/%d")
        if sd < date < ed:
            publish = msg.select("a")[2].text
            c = c + 1
            for i in range(c-1,c):
                pd[i] = publish
    mid_listpd = list(pd.values())
    listpd = list(set(mid_listpd))
    for l in listpd:
        C.append(mid_listpd.count(l))
    for i in range (len(listpd)):
        listdata = [listpd[i],C[i]]
        list2.append(listdata)
    print(list2)
```

```python
def showmdata(url,page):
    html = requests.get(url, headers = headers).text
    soup = BeautifulSoup(html,"lxml")
    res = soup.find_all("div",{"class":"mod type02_m012 clearfix"})[0]
    items = res.select(".item")
    c1 = c2 = c3 = 0
    slist = elist =[]
    FS = datetime.strptime("2022/2/1","%Y/%m/%d")
    MS = datetime.strptime("2022/3/1","%Y/%m/%d")
    AS = datetime.strptime("2022/4/1","%Y/%m/%d")
    FE = datetime.strptime("2022/2/28","%Y/%m/%d")
    ME = datetime.strptime("2022/3/31","%Y/%m/%d")
    AE = datetime.strptime("2022/4/30","%Y/%m/%d")
    month = ["2月","3月","4月"]
    for item in items:
#     item = items[0]
        msg = item.select(".msg")[0]
        sd = datetime.strptime('2022/2/1',"%Y/%m/%d")
        ed = datetime.strptime('2022/4/30',"%Y/%m/%d")
        mid_date = msg.find("span").text.split("：")[-1]
        date = datetime.strptime(mid_date,"%Y/%m/%d")
        if sd < date < ed:
            if FS < date < FE:
                c1 += 1
            elif MS < date < ME:
                c2 += 1
            else:
                c3 += 1
        C = [c1,c2,c3]
    for i in range(len(C)):
        listdata = [month[i],C[i]]
        list3.append(listdata)
    print(list3)
```
